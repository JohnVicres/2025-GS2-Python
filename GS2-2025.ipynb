```python
#João Vicente RM556561

import random as ran
tarefas=[]

#simulação para adicionar 50 tarefas de pesos aleatórios na fila de tarefas
def construcaoInicialFila(tarefas):
    while len(tarefas)<50:                  #LEMBRETE: trocar por 500 na entrega
        n=ran.randrange(1,20,1)
        tarefas.append(n)
    return tarefas

#quick sort das tarefas por peso
def quickSortFila(tarefas, d=0):
    if len(tarefas)<=1:
        return tarefas
    pivo=tarefas[-1]
    menores=[x for x in tarefas[:-1] if x<=pivo]
    maiores=[x for x in tarefas[:-1] if x>pivo]
    return quickSortFila(menores,d+1)+[pivo]+quickSortFila(maiores,d+1)


#quickSortFila(construcaoInicialFila(tarefas))
```

```python
import timeit as t
#"área" de processamento de itens (tarefas individuais) pela IA
def atendFilaIA(item):
    return item
    
#"área" de ~~ pelo humano
def atendFilaHum(item):
    return item
    
#processamento da fila

def processarFila(cont=0):
    filaord=quickSortFila(tarefas)
    if len(filaord)<=1:
        print("Fila de tarefas terminada")
        return
    itemIA=filaord.pop(0)
    itemHum=filaord.pop(len(filaord)-1)
    t.Timer(1000+500*itemIA, atendFilaIA(itemIA))
    if len(filaord)>=1:
        t.Timer(5000+200*itemHum, atendFilaHum(itemHum))
    cont+=1
    if cont%4 and len(filaord)<=2:
        processarFila(cont)
    

    
```
