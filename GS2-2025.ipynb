```python
#João Vicente RM556561

import random as ran
tarefas=[]

#simulação para adicionar 50 tarefas de pesos aleatórios na fila de tarefas
def construcaoInicialFila(tarefas):
    while len(tarefas)<50:                  #LEMBRETE: trocar por 500 na entrega
        n=ran.randrange(1,20,1)
        tarefas.append(n)
    return tarefas

#quick sort das tarefas por peso
def quickSortFila(tarefas, d=0):
    if len(tarefas)<=1:
        return tarefas
    pivo=tarefas[-1]
    menores=[x for x in tarefas[:-1] if x<=pivo]
    maiores=[x for x in tarefas[:-1] if x>pivo]
    return quickSortFila(menores,d+1)+[pivo]+quickSortFila(maiores,d+1)


#quickSortFila(construcaoInicialFila(tarefas))
```

```python
import time as t
#"área" de processamento de itens (tarefas individuais) pela IA
def atendFilaIA(item):
    return item
    
#"área" de ~~ pelo humano
def atendFilaHum(item):
    return item
    
#processamento da fila
def processarFila(fila,cont=0):
    ###SETUP###
    
    #match-case para fim da recursão, ou para caso a fila tenha apenas uma tarefa sobrando
    match len(fila):
        case 0:
            print(f"Fila de tarefas terminada!\nTotal de tarefas cumpridas: {cont}")
            return
        case 1:
            itemfinal=fila.pop(0)
            if 1+0.5*itemfinal>=5+0.1*itemfinal:
                t.sleep(1+0.5*itemfinal)
                atendFilaIA(itemfinal)
                return
            else:
                t.sleep(5+0.1*itemfinal)
                atendFilaHum(itemfinal)
                return
    
    #organização da fila
    filaord=quickSortFila(fila)
    
    #itens a serem removidos da fila e processados
    itemIA=filaord.pop(0)
    itemHum=filaord.pop(len(filaord)-1)
    
    ###ATENDIMENTO###
    #1s+0.5s*peso para IA
    t.sleep(1+0.5*itemIA)
    atendFilaIA(itemIA)
    #5s+0.1*peso para humano
    t.sleep(5+0.1*itemHum)
    atendFilaHum(itemHum)

    
    ###SAÍDAS###
    #se já se passaram 50 atendimentos, imprimir info da fila
    if cont>1 and len(filaord)%50:
        print(f"Comprimento da fila de tarefas: {len(filaord)}\nPeso total da fila de tarefas: {sum(filaord)}\n")
        
    #recursão
    return processarFila(filaord,cont+1)
```
