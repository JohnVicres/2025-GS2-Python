```python
#João Vicente RM556561

import random as ran
import time as t

tarefas=[]

#simulação para adicionar 50 tarefas de pesos aleatórios na fila de tarefas
def construcaoInicialFila(tarefas):
    while len(tarefas)<10:
        n=ran.randrange(1,20,1)
        tarefas.append(n)
    return tarefas

#quick sort das tarefas por peso
def quickSortFila(tarefas, d=0):
    if len(tarefas)<=1:
        return tarefas
    pivo=tarefas[-1]
    menores=[x for x in tarefas[:-1] if x<=pivo]
    maiores=[x for x in tarefas[:-1] if x>pivo]
    return quickSortFila(menores,d+1)+[pivo]+quickSortFila(maiores,d+1)


historico_atendimento={"IA":0,"Humano":0}

#atendimento de itens pela IA
def atendFilaIA(item):
    historico_atendimento["IA"]+=1
    h=historico_atendimento["IA"]
    print(f"IA processou seu {h}o item, de peso {item}")
    
#atendimento de itens pelo Humano
def atendFilaHum(item):
    historico_atendimento["Humano"]+=1
    h=historico_atendimento["Humano"]
    print(f"Humano processou seu {h}o item, de peso {item}")
    
#processamento da fila
def processarFila(fila,total_atendido=0):
    ###SETUP###
    
    #match-case para fim da recursão, ou para caso a fila tenha apenas uma tarefa sobrando
    match len(fila):
        case 0:
            print(f"Fila de tarefas terminada!\nTotal de tarefas cumpridas: {total_atendido}")
            return
        case 1:
            itemfinal=fila.pop(0)
            if 0.1+0.5*itemfinal>=0.5+0.1*itemfinal:
                t.sleep(0.1+0.5*itemfinal)
                atendFilaIA(itemfinal)
                return
            else:
                t.sleep(0.5+0.1*itemfinal)
                atendFilaHum(itemfinal)
                return
    
    #organização da fila
    filaord=quickSortFila(fila)
    
    #itens a serem removidos da fila e processados
    itemIA=filaord.pop(0)
    itemHum=filaord.pop(len(filaord)-1)
    
    ###ATENDIMENTO###
    
    #0.1s+0.5s*peso para IA
    t.sleep(0.1+0.5*itemIA)
    atendFilaIA(itemIA)
    #0.5s+0.1s*peso para humano
    t.sleep(0.5+0.1*itemHum)
    atendFilaHum(itemHum)
    total_atendido=historico_atendimento["IA"]+historico_atendimento["Humano"]
    
    ###SAÍDA###
    
    #se já se passaram 10 atendimentos, imprimir info da fila
    if len(filaord)%5==0:
        print(f"\nTotal atendido: {total_atendido}\nTarefas restantes: {len(filaord)}\nPeso total restante: {sum(filaord)}\n")
        
    #recursão
    return processarFila(filaord,total_atendido)


construcaoInicialFila(tarefas)
processarFila(tarefas)
```
